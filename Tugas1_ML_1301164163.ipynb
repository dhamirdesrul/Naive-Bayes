```python
#Naive Bayes
#membuat array baru dengan dengan ukuran yang sesuai dengan banyaknya 40 (meneyesuaikan dengan isi dalam data testset) indeks yang menyimpan hasil data set yang akan diberikan label
array_total = np.ones((40,2))
variable_iterasi = 0

for i in range(len(dFT)):
    for j in range(1, 8):
        variable_0 = arrayKosong[j-1][dFT[i][j]][0]/arrayBaru[j-1][3][0] #kelas 0, menyimpan label kelas 0 
        variable_1 = arrayKosong[j-1][dFT[i][j]][1]/arrayBaru[j-1][3][1] #kelas 1, menyimpan label kelas 1
        array_total[variable_iterasi][0]*= variable_0  #mengalikan ke dalam array_total sacara keseluruhan dengan variable 0
        array_total[variable_iterasi][1]*= variable_1 #mengalikan ke dalam array_total sacara keseluruhan dengan variable 0
    array_total[variable_iterasi][0]*= 120/160 #variable 0 memiliki totak yakni <=50K jadi akan dihitung probabilitasnya dengan keseluruhan banyaknya <=50k dengan pembandingnya jumlah keseluruhan label kelas
    array_total[variable_iterasi][1]*= 40/160 #variable 1 memiliki totak yakni >50K jadi akan dihitung probabilitasnya dengan keseluruhan banyaknya >50k dengan pembandingnya jumlah keseluruhan label kelas
    variable_iterasi += 1
    
    
print(array_total)
```

```python
# mengetahui nilai maksimum tapi keluarannya yakni indeks yang memiliki nilai terbesar
a = np.argmax(array_total, axis=1)
```

```python
hasil = pandas.DataFrame(a).replace({0: '>50K', 1: '<=50K'})

hasil
```
